```python
# Data
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

#warning
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('train_dataset2.csv')
dataset
```

```python
dataset.info()
```

```python
dataset.drop(['Unnamed: 0'], axis=1, inplace = True)
```

```python
corr_data = dataset.corr()
corr_data = corr_data.apply(lambda x: round(x ,2))
corr_data
```

```python
corr_data.to_csv('corr_data.csv')
```

```python
s = corr_data.unstack()
s
```

```python
# 변수간의 상관관계
df = pd.DataFrame(s[s < 1].sort_values(ascending=False), columns=['corr'])
df.style.background_gradient(cmap='viridis')
```
