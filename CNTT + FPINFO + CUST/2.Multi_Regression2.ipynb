## 유의하지 않은 변수 제거 후

```python
#모듈IMPORT
import pandas as pd
import statsmodels.api as sm
```

```python
#데이터 불러오기
train_dataset=pd.read_csv('processed_CNTT_FPINFO_CUST.csv')
```

```python
train_dataset.head()
```

```python
#DIVIDED_SET 2인 행 제거
train_dataset=train_dataset[train_dataset.DIVIDED_SET != 2]
```

```python
train_dataset.info()
```

```python
#DIVIDED_SET, CUST_ID, CLLT_FP_PRNO제거
train_dataset.drop(['DIVIDED_SET'], axis=1, inplace = True)
train_dataset.drop(['CUST_ID'], axis=1, inplace = True)
train_dataset.drop(['CLLT_FP_PRNO'], axis=1, inplace = True)
```

```python
#SIU_CUST_YN Y/N ==> 1/0으로 변경
train_dataset['SIU_CUST_YN'].replace('Y', 1, inplace = True)
train_dataset['SIU_CUST_YN'].replace('N', 0, inplace = True)
```

```python
#FP_CAREER Y/N ==> 1/0으로 변경
train_dataset['FP_CAREER'].replace('Y', 1, inplace = True)
train_dataset['FP_CAREER'].replace('N', 0, inplace = True)
```

```python
#WEDD_YN Y/N ==> 1/0으로 변경
train_dataset['WEDD_YN'].replace('Y', 1, inplace = True)
train_dataset['WEDD_YN'].replace('N', 0, inplace = True)
```

```python
#INCB_DVSN ==> R : 1. P : 2, C : 3
train_dataset['INCB_DVSN'].replace('R', 1, inplace = True)
train_dataset['INCB_DVSN'].replace('P', 2, inplace = True)
train_dataset['INCB_DVSN'].replace('C', 3, inplace = True)
```

```python
train_dataset.info()
```

```python
#유의하지 않는 변수 제거
train_dataset.drop(['INCB_DVSN'], axis=1, inplace = True)
train_dataset.drop(['WEDD_YN'], axis=1, inplace = True)
train_dataset.drop(['MAX_PRM'], axis=1, inplace = True)
train_dataset.drop(['RGST_YEAR'], axis=1, inplace = True)
train_dataset.drop(['RGST_MONTH'], axis=1, inplace = True)
train_dataset.drop(['MNTH_INCM_AMT_AVG'], axis=1, inplace = True)
train_dataset.drop(['SUM_ORIG_PREM_SUM'], axis=1, inplace = True)
train_dataset.drop(['CNTT_TERM_AVG'], axis=1, inplace = True)
train_dataset.drop(['WORK_YEARS_MAX'], axis=1, inplace = True)
```

```python
train_dataset.head()
```

```python
train_dataset.info()
```

```python
train_dataset.to_csv('train_dataset2.csv')
```

```python
prediction = train_dataset[['SIU_CUST_YN']] # y : 보험사기여부
```

```python
# x1, x2, ... 나머지 변수
train_data = train_dataset[['SEX','AGE','FP_CAREER','OCCP_GRP','TOTALPREM','MAIN_INSR_AMT_SUM','EXPR_SUM','WORK_YEARS_MIN','EXPR_COUNT']]
train_data
```

```python
train_data = sm.add_constant(train_data, has_constant = 'add')
```

```python
multi_model = sm.OLS(prediction, train_data)
fitted_multi_model = multi_model.fit()
```
