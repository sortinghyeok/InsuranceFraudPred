### 데이터 변환

##### xlsx to csv   
##### csv to csv   


- 구분자 쉼표
- 인코딩 방식 UTF-8
- 결측치 NaN

```python
import pandas as pd
```

CNTT데이터 CONVERT

```python
data = pd.read_csv('CNTT.csv')
data.head()
```

```python
data.to_csv('CNTTconvert.csv',sep=',',na_rep='',encoding='UTF-8',index=False)
```
