```python
#모듈IMPORT
import pandas as pd
```

```python
dataset=pd.read_csv('CNTT_FPINFO_CUST.csv')
dataset.head()
```

```python
dataset.info()
```

```python
# 의미없는 column 제거
dataset.drop(['POLY_NO'], axis=1, inplace = True)
dataset.drop(['SALE_CHNL_CODE'], axis=1, inplace = True)
dataset.drop(['CNTT_STAT_CODE'], axis=1, inplace = True)
dataset.drop(['EXTN_YM'], axis=1, inplace = True)
dataset.drop(['LAPS_YM'], axis=1, inplace = True)
dataset.drop(['PAYM_CYCL_CODE'], axis=1, inplace = True)
dataset.drop(['BRCH_CODE'], axis=1, inplace = True)
```

```python
dataset.info()
```

```python
#CUST_ID기준 MNTH_INCM_AMT 평균처리
MNTH=dataset.groupby('CUST_ID')['MNTH_INCM_AMT'].mean().rename('MNTH_INCM_AMT_avg').reset_index()
dataset=dataset.merge(MNTH)
```

```python
dataset
```

```python
#MNTH_INCM_AMT제거
dataset.drop(['MNTH_INCM_AMT'], axis=1, inplace = True)
```

```python
#CUST_ID기준 MAIN_INSR_AMT 합계처리
main=dataset.groupby('CUST_ID')['MAIN_INSR_AMT'].sum().rename('MAIN_INSR_AMT_sum').reset_index()
dataset=dataset.merge(main)
```

```python
#CUST_ID기준 SUM_ORIG_PREM 합계처리
orig=dataset.groupby('CUST_ID')['SUM_ORIG_PREM'].sum().rename('SUM_ORIG_PREM_sum').reset_index()
dataset=dataset.merge(orig)
```

```python
dataset
```

```python
#MAIN_INSR_AMT, SUM_ORIG_PREM 제거
dataset.drop(['MAIN_INSR_AMT'], axis=1, inplace = True)
dataset.drop(['SUM_ORIG_PREM'], axis=1, inplace = True)
```

```python
dataset.info()
```

```python
dataset.to_csv('dataset.csv')
```
