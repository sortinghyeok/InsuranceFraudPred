```python
import pandas as pd
```

```python
import math
b = math.nan

def isNaN(num):
    return num != num
print(isNaN(b))
```

```python
custDF = pd.read_csv("../data_convert/CNTTconvert.csv",  sep=',', encoding = 'utf-8')
```

```python
custDF = custDF.drop(custDF.index[0])
```

```python
custDF
```

```python
custDF.info()
```

```python
#결측치 개수 확인
custDF.isnull().sum()
# CLLT_FP_PRNO, CNTT_STAT_CODE, CNTT_STAT_CODE 등 결측치 값이 같은 column들이 존재 하기 때문에 
# 같은 row 일 가능성이 있음.
```

```python
# 우선 FPINFO 값이 없는 row를 기준으로 제거
custDF.dropna(subset = ['CLLT_FP_PRNO'], inplace= True)
```

```python
custDF.isnull().sum()
# 대다수의 결측치 제거 확인
```

```python
custDF.dropna(subset = ['RECP_PUBL'], inplace= True)
```

```python
custDF.isnull().sum()
```

```python
# 의미없는 column 제거
custDF.drop(['IRKD_CODE_DTAL'], axis=1, inplace = True)
custDF.drop(['IRKD_CODE_ITEM'], axis=1, inplace = True)
```

```python
custDF.isnull().sum()
```

```python
custDF['EXPR_YM'].replace(999912, int(1), inplace = True)
```

```python
custDF
```

```python
print(type(custDF['EXPR_YM'][1]))
```

```python
# 만기년월 값, 종신여부로 구분 (종신의 경우 1 아니면 0)
custDF['EXPR_YM'] = custDF['EXPR_YM'].apply(lambda x: 1 if x==1 else 0)
```

```python
custDF
```

```python
# 소멸년월 값 존재 유무로 구분 (소멸한 적이 있으면 1 없으면 0)
custDF['EXTN_YM'] = custDF['EXTN_YM'].apply(lambda x: 0 if isNaN(x) else 1)
```

```python
custDF
```

```python
# 실효년월 값 존재 유무로 구분 (실효된 적이 있으면 1 없으면 0)
custDF['LAPS_YM'] = custDF['LAPS_YM'].apply(lambda x: 0 if isNaN(x) else 1)
```

```python
custDF
```

```python

# 계약 상태코드 알파벳과 숫자가 혼재되어있어 전체 다 숫자로 변환 

#순서대로 
#유지(1),완납(2),납입면제(3),효력상실(4),만기경과(5)
#만기지급(6),해약(7),사망감소(8),장해감소(9),해지 (10)
#무효(11),정산(12),청약철회(13),반송(14),부활취소(15)

custDF['CNTT_STAT_CODE'].replace(9, 5, inplace = True)

custDF['CNTT_STAT_CODE'].replace('A', 6, inplace = True)

custDF['CNTT_STAT_CODE'].replace('B', 7, inplace = True)

custDF['CNTT_STAT_CODE'].replace('C', 8, inplace = True)

custDF['CNTT_STAT_CODE'].replace('D', 9, inplace = True)

custDF['CNTT_STAT_CODE'].replace('E', 10, inplace = True)

custDF['CNTT_STAT_CODE'].replace('G', 11, inplace = True)

custDF['CNTT_STAT_CODE'].replace('H', 12, inplace = True)

custDF['CNTT_STAT_CODE'].replace('I', 13, inplace = True)

custDF['CNTT_STAT_CODE'].replace('J', 14, inplace = True)

custDF['CNTT_STAT_CODE'].replace('L', 15, inplace = True)
```

```python
custDF
```

```python
custDF.drop(['DISTANCE'], axis=1, inplace = True)
```
