```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')
```

```python
fpinfoDF = pd.read_csv("FPINFOconvert.csv",  sep=',', encoding = 'utf-8')
```

```python
fpinfoDF
```

```python
fpinfoDF.info()
```

```python
#결측치 개수 확인
fpinfoDF.isnull().sum()
```

```python
# INCB_DVSN 변수의'D', '#' 값인 전체 row 제거
idx = fpinfoDF[(fpinfoDF['INCB_DVSN'] == 'D') | (fpinfoDF['INCB_DVSN'] == '#')].index                                                 
fpinfoDF = fpinfoDF.drop(idx)
fpinfoDF
```

```python
fpinfoDF.isnull().sum() # 결측치 2개씩 존재
```

```python
fpinfoDF = fpinfoDF.dropna() # 변수별로 결측치 제거
```

```python
fpinfoDF.isnull().sum() #결측치 제거 확인
```

```python
# BEFO_JOB 변수에 '보험관계인'일 경우 1, 그렇지 않은 경우는 0으로 하여 보험관련 업무 경험 여부
fpinfoDF['BEFO_JOB'].replace('보험관계인', 1, inplace = True)

fpinfoDF['BEFO_JOB'].replace('교육공무원', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('공무원', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('교육부문', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('국영기업체', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('군인', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('금융부문', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('기타', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('무직', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('사기업체', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('영업직(판', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('의료부문', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('일반공무원', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('일반기업', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('자영업', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('자유업', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('학생', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('회사원', 0, inplace = True)
fpinfoDF['BEFO_JOB'].replace('#', 0, inplace = True)
```

```python
fpinfoDF['BEFO_JOB']
```

```python
# EDGB 변수의 '#' 값은 FP(보험설계사)의 최종학력을 알 수 없는 상태로 '기타' 로 설정
fpinfoDF['EDGB'].replace('#', '기타', inplace = True)
```

```python
fpinfoDF['EDGB']
```

```python
fpinfoDF.isnull().sum() # 결측치가 없는지 다시 최종 확인
```
